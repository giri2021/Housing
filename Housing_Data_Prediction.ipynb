# DataPrediction

```python
# list of variable to map 
variable =['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
def binary_map(x):
    return x.map({'yes':1,'no':0})
housing[variable]= housing[variable].apply(binary_map)
```

```python
housing.head()
```

```python
# Dummy varibble
status = pd.get_dummies(housing['furnishingstatus'])
status.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'],drop_first =True)
housing =pd.concat([housing,status],axis =1)
housing.head()
```

```python
housing.drop(['furnishingstatus'],axis=1 ,inplace=True)
housing.head()
```

# splitting the  Data into Train and Test

```python
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(housing,train_size=0.7,test_size=0.3,random_state =100)
```

# Rescaling the Feature

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
num_var = ['area','bedrooms','bathrooms','stories','parking','price']
df_train[num_var] = scaler.fit_transform(df_train[num_var])
df_train.head()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

# Building model
- we will using the LinearRegression  function from scikit Learn fro compatibility with RFE

# RFE
- Recursive feature elemation

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm =LinearRegression ()
lm.fit(X_train,y_train)
rfe = RFE(lm,n_features_to_select =10)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

# Building model using statamodel for the detailed statistics

```python
# creating x test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe) 
```

```python
lm =sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
X_train_new = X_train_rfe.drop(["bedrooms"], axis = 1)
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm =sm.OLS(y_train,X_train_lm).fit()
```

```python
print(lm.summary())
```

```python
X_train_new =X_train_new.drop(["const"],axis =1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif =pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF']= [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF']= round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF",ascending = False)
vif
```

```python
y_train_pre =  lm.predict(X_train_lm)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_pre),bins =20)
fig.suptitle("Error Terms",fontsize = 20)
plt.xlabel("Errors",fontsize =18)
```
